```python
import tweepy
import TwitterCredentials as keys
import pandas as pd

auth = tweepy.OAuthHandler(keys.consumer_API_key, keys.consumer_API_secret_key)
auth.set_access_token(keys.access_token, keys.access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

#cursor objects for each news source
wsj = tweepy.Cursor(api.user_timeline, screen_name='wsj', tweet_mode='extended')
cnn = tweepy.Cursor(api.user_timeline, screen_name='CNN', tweet_mode='extended')
nyt = tweepy.Cursor(api.user_timeline, screen_name='nytimes', tweet_mode='extended')
fox = tweepy.Cursor(api.user_timeline, screen_name='foxnews', tweet_mode='extended')

#define empty lists to hold each tweet
wsjlist = []
cnnlist = []
nytlist = []
foxlist = []

#store each tweet in the cursor object in the corresponding list
for tweet in wsj.items(900):
    wsjlist.append(tweet)

for tweet in cnn.items(900):
    cnnlist.append(tweet)
    
for tweet in nyt.items(900):
    nytlist.append(tweet)
    
for tweet in fox.items(900):
    foxlist.append(tweet)
    
#store data as a data frame
df = pd.DataFrame({"Wall Street Journal": wsjlist, "CNN": cnnlist, "New York Times": nytlist, "Fox": foxlist})
```

```python
#function to convert the tweepy object to text
def totext(tweet):
    return tweet.full_text

#define new data frame to hold texts of tweets
dfnew = pd.DataFrame({"Wall Street Journal": [], "CNN": [], "New York Times": [], "Fox": []})

#map above function to each column in df
dfnew["Wall Street Journal"] = df["Wall Street Journal"].map(totext)
dfnew["CNN"] = df["CNN"].map(totext)
dfnew["New York Times"] = df["New York Times"].map(totext)
dfnew["Fox"] = df["Fox"].map(totext)
```
